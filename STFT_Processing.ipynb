```python
#Mengambil File dengan Ekstensi CSV pada Setiap Folder Kondisi Gear
healthy_file = glob.glob(cur_path+'/Healthy/*.csv')
brokentooth_file = glob.glob(cur_path+'/BrokenTooth/*.csv')
```

```python
# Load Data
# Function data healty direction a1
def data_x1(healthy_file):
    datax1 = pd.DataFrame()
    for fx1 in healthy_file:
        dfx1 = pd.read_csv(fx1, usecols = [0], header=None) # read the csv file
        datax1 = pd.concat([datax1,dfx1],axis=1,ignore_index=True)
    return datax1

# Function data healty direction a2
def data_x2(healthy_file):
    datax2 = pd.DataFrame()
    for fx2 in healthy_file:
        dfx2 = pd.read_csv(fx2, usecols = [1], header=None) # read the csv file
        datax2 = pd.concat([datax2,dfx2],axis=1,ignore_index=True)
    return datax2

# Function data healty direction a3
def data_x3(healthy_file):
    datax3 = pd.DataFrame()
    for fx3 in healthy_file:
        dfx3 = pd.read_csv(fx3, usecols = [2], header=None) # read the csv file
        datax3 = pd.concat([datax3,dfx3],axis=1,ignore_index=True)
    return datax3

# Function data healty direction a4
def data_x4(healthy_file):
    datax4 = pd.DataFrame()
    for fx4 in healthy_file:
        dfx4 = pd.read_csv(fx4, usecols = [3], header=None) # read the csv file
        datax4 = pd.concat([datax4,dfx4],axis=1,ignore_index=True)
    return datax4

# Function data broken tooth direction a1
def data_y1(brokentooth_file):
    datay1 = pd.DataFrame()
    for fy1 in brokentooth_file:
        dfy1 = pd.read_csv(fy1, usecols = [0], header=None) # read the csv file
        datay1 = pd.concat([datay1,dfy1],axis=1,ignore_index=True)
    return datay1

# Function data broken tooth direction a2
def data_y2(brokentooth_file):
    datay2 = pd.DataFrame()
    for fy2 in brokentooth_file:
        dfy2 = pd.read_csv(fy2, usecols = [1], header=None) # read the csv file
        datay2 = pd.concat([datay2,dfy2],axis=1,ignore_index=True)
    return datay2

# Function data broken tooth direction a3
def data_y3(brokentooth_file):
    datay3 = pd.DataFrame()
    for fy3 in brokentooth_file:
        dfy3 = pd.read_csv(fy3, usecols = [2], header=None) # read the csv file
        datay3 = pd.concat([datay3,dfy3],axis=1,ignore_index=True)
    return datay3

# Function data broken tooth direction a4
def data_y4(brokentooth_file):
    datay4 = pd.DataFrame()
    for fy4 in brokentooth_file:
        dfy4 = pd.read_csv(fy4, usecols = [3], header=None) # read the csv file
        datay4 = pd.concat([datay4,dfy4],axis=1,ignore_index=True)
    return datay4
```

```python
#Data Healthy Tiap Arah a1, a2, a3, a4
data_healthy_1 = data_x1(healthy_file).T
data_healthy_2 = data_x2(healthy_file).T
data_healthy_3 = data_x3(healthy_file).T
data_healthy_4 = data_x4(healthy_file).T
```

```python
#Data Broken Tooth Tiap Arah a1, a2, a3, a4
data_broken_1 = data_y1(brokentooth_file).T
data_broken_2 = data_y2(brokentooth_file).T
data_broken_3 = data_y3(brokentooth_file).T
data_broken_4 = data_y4(brokentooth_file).T
```

**Exploratory Data Analysis (EDA)**

```python
import matplotlib.pyplot as plt
import os

dir_path1 = cur_path+'/Healthy'
dir_path2 = cur_path+'/BrokenTooth'

data_lengths = [len([entry for entry in os.listdir(dir_path1) if os.path.isfile(os.path.join(dir_path1, entry))]),
                len([entry for entry in os.listdir(dir_path2) if os.path.isfile(os.path.join(dir_path2, entry))])]

labels = ['Healthy', 'Broken Tooth']
percentages = ['{:.1f}%'.format((length / sum(data_lengths)) * 100) for length in data_lengths]
data_labels = [f'{percentage}\n({length} Data)' for percentage, length in zip(percentages, data_lengths)]

fig, ax = plt.subplots()
patches, texts, autotexts = ax.pie(data_lengths, labels=labels, autopct='',  # Menghilangkan persentase default dan label
                                   wedgeprops={'edgecolor': 'black'}, counterclock=False, shadow=True, startangle=0,
                                   radius=1.3, textprops={'fontsize': 12, 'weight': 'bold'})

# Menambahkan teks persentase dan jumlah data asli ke dalam lingkaran pie
for i, autotext in enumerate(autotexts):
    autotext.set_text(data_labels[i])  # Menggunakan data_labels yang sudah berisi persentase dan jumlah data asli

plt.tight_layout()
plt.legend(labels, loc=0, bbox_to_anchor=(1, 1), fontsize=12)
plt.show()
```

```python
#Crest Factor
Crestf_healthy_1 = pd.DataFrame(np.real(np.max(np.abs(healthy_1), axis=(1, 2)) / np.sqrt(np.mean(healthy_1 ** 2, axis=(1, 2)))))
Crestf_healthy_2 = pd.DataFrame(np.real(np.max(np.abs(healthy_2), axis=(1, 2)) / np.sqrt(np.mean(healthy_2 ** 2, axis=(1, 2)))))
Crestf_healthy_3 = pd.DataFrame(np.real(np.max(np.abs(healthy_3), axis=(1, 2)) / np.sqrt(np.mean(healthy_3 ** 2, axis=(1, 2)))))
Crestf_healthy_4 = pd.DataFrame(np.real(np.max(np.abs(healthy_4), axis=(1, 2)) / np.sqrt(np.mean(healthy_4 ** 2, axis=(1, 2)))))

Crestf_broken_1 = pd.DataFrame(np.real(np.max(np.abs(broken_1), axis=(1, 2)) / np.sqrt(np.mean(broken_1 ** 2, axis=(1, 2)))))
Crestf_broken_2 = pd.DataFrame(np.real(np.max(np.abs(broken_2), axis=(1, 2)) / np.sqrt(np.mean(broken_2 ** 2, axis=(1, 2)))))
Crestf_broken_3 = pd.DataFrame(np.real(np.max(np.abs(broken_3), axis=(1, 2)) / np.sqrt(np.mean(broken_3 ** 2, axis=(1, 2)))))
Crestf_broken_4 = pd.DataFrame(np.real(np.max(np.abs(broken_4), axis=(1, 2)) / np.sqrt(np.mean(broken_4 ** 2, axis=(1, 2)))))

Crestf_healthy = pd.concat([Crestf_healthy_1,Crestf_healthy_2,Crestf_healthy_3,Crestf_healthy_4], axis=1,ignore_index=True)
Crestf_broken = pd.concat([Crestf_broken_1,Crestf_broken_2,Crestf_broken_3,Crestf_broken_4], axis=1,ignore_index=True)

#Variance
variance_healthy_1 = pd.DataFrame(np.var(np.abs(healthy_1), axis=(1, 2)))
variance_healthy_2 = pd.DataFrame(np.var(np.abs(healthy_2), axis=(1, 2)))
variance_healthy_3 = pd.DataFrame(np.var(np.abs(healthy_3), axis=(1, 2)))
variance_healthy_4 = pd.DataFrame(np.var(np.abs(healthy_4), axis=(1, 2)))

variance_broken_1 = pd.DataFrame(np.var(np.abs(broken_1), axis=(1, 2)))
variance_broken_2 = pd.DataFrame(np.var(np.abs(broken_2), axis=(1, 2)))
variance_broken_3 = pd.DataFrame(np.var(np.abs(broken_3), axis=(1, 2)))
variance_broken_4 = pd.DataFrame(np.var(np.abs(broken_4), axis=(1, 2)))

variance_healthy = pd.concat([variance_healthy_1,variance_healthy_2,variance_healthy_3,variance_healthy_4], axis=1,ignore_index=True)
variance_broken = pd.concat([variance_broken_1,variance_broken_2,variance_broken_3,variance_broken_3], axis=1,ignore_index=True)

#Form Factor
FF_healthy_1 = pd.DataFrame(np.real(np.sqrt(np.mean(healthy_1 ** 2, axis=(1, 2))) / np.mean(np.abs(healthy_1), axis=(1, 2))))
FF_healthy_2 = pd.DataFrame(np.real(np.sqrt(np.mean(healthy_2 ** 2, axis=(1, 2))) / np.mean(np.abs(healthy_2), axis=(1, 2))))
FF_healthy_3 = pd.DataFrame(np.real(np.sqrt(np.mean(healthy_3 ** 2, axis=(1, 2))) / np.mean(np.abs(healthy_3), axis=(1, 2))))
FF_healthy_4 = pd.DataFrame(np.real(np.sqrt(np.mean(healthy_4 ** 2, axis=(1, 2))) / np.mean(np.abs(healthy_4), axis=(1, 2))))

FF_broken_1 = pd.DataFrame(np.real(np.sqrt(np.mean(broken_1 ** 2, axis=(1, 2))) / np.mean(np.abs(broken_1), axis=(1, 2))))
FF_broken_2 = pd.DataFrame(np.real(np.sqrt(np.mean(broken_2 ** 2, axis=(1, 2))) / np.mean(np.abs(broken_2), axis=(1, 2))))
FF_broken_3 = pd.DataFrame(np.real(np.sqrt(np.mean(broken_3 ** 2, axis=(1, 2))) / np.mean(np.abs(broken_3), axis=(1, 2))))
FF_broken_4 = pd.DataFrame(np.real(np.sqrt(np.mean(broken_4 ** 2, axis=(1, 2))) / np.mean(np.abs(broken_4), axis=(1, 2))))


FF_healthy = pd.concat([FF_healthy_1,FF_healthy_2,FF_healthy_3,FF_healthy_4], axis=1,ignore_index=True)
FF_broken = pd.concat([FF_broken_1,FF_broken_2,FF_broken_3,FF_broken_4], axis=1,ignore_index=True)
```
