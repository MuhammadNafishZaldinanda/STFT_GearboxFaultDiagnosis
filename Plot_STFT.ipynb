```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.signal import stft

df = pd.read_csv(r'D:\python\code\Gear\Data\Healthy\h30hz0.csv')
data_healthy_1 = df.iloc[:,2].values  # Ganti 'column_name' dengan nama kolom yang berisi data

fs = 30  # Frekuensi sampel

# Hitung STFT
frequencies, times, healthy_1 = stft(data_healthy_1, fs=fs, window='hann', nperseg=256, noverlap=None)

# Tampilkan spektrum
plt.figure(figsize=(15,7))
plt.pcolormesh(times, frequencies, np.abs(healthy_1), shading='auto', cmap='jet')
plt.colorbar(label='Amplitudo')
plt.ylabel('Frekuensi (Hz)')
plt.xlabel('Waktu (detik)')
plt.title('Spektrum STFT (Helthy Gear)')
plt.ylim(0, 15)  # Batas frekuensi yang akan ditampilkan
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.signal import stft

df = pd.read_csv(r'D:\python\code\Gear\Data\BrokenTooth\b30hz0.csv')
waktu = df.iloc[:,0]  
arah1 = df.iloc[:, 0].values  
arah2 = df.iloc[:, 1].values  
arah3 = df.iloc[:, 2].values  
arah4 = df.iloc[:, 2].values  

fs = 30  # Frekuensi sampel

# Hitung STFT
frequencies, times, healthy_1 = stft(arah1, fs=fs, window='hann', nperseg=256, noverlap=None)

f = 30
t = [(i / f) for i in range(len(waktu))] 
t = pd.DataFrame(t, columns=[None])


plt.figure(figsize=(20,5))
plt.plot(t, arah1, color = "blue", )
#plt.plot(t, arah2, color = "green")
#plt.plot(t, arah3, color = "red")
#plt.plot(t, arah4, color = "yellow")
plt.xlabel('Time (s)',fontsize=16)
plt.ylabel('Acceleration (G)', fontsize=16)
plt.title('Time Domain (Broken Tooth)')
plt.show()

from numpy.fft import rfft, rfftfreq
x = waktu
#x.drop(x.columns[[1,2]], axis=1, inplace=True)
x=x.values.flatten() #to convert DataFrame to 1D array
x=x*9.81
# Calculate Fast Fourier Transform
x = np.asarray(x)
n=len(x)
fs=30 # Example sampling frequency in Hz
dt=1/fs #time increment in each data
xfftx = rfft(x)*dt
xfftx = abs(xfftx)
xfftx = (xfftx - np.min(xfftx)) / (22 - np.min(xfftx))
fAxis = rfftfreq(n,dt) # in Hz
plt.figure(figsize=(20,5))
plt.plot(fAxis,xfftx, color = "blue")
plt.xlim(0,15)
plt.ylim(-0.5,710)
plt.xlabel('Frekuensi [Hz]')
plt.ylabel('Acceleration (G)')
plt.title('Spektrum FFT (Broken Tooth)')
plt.show()

plt.figure(figsize=(15,7))
plt.pcolormesh(times, frequencies, np.abs(healthy_1), shading='auto', cmap='jet')
plt.colorbar(label='Amplitudo')
plt.ylabel('Frekuensi (Hz)')
plt.xlabel('Time (s)')
plt.title('Spektrum STFT (Broken Tooth)')
plt.ylim(0, 15)  # Batas frekuensi yang akan ditampilkan
plt.show()
```
